```python
df=pd.read_csv("dt.csv")
df=df.dropna()

df
```

```python
inputs = df.drop('income',axis='columns')
target = df['income']
```

```python
from sklearn.preprocessing import LabelEncoder
workclass = LabelEncoder()
education = LabelEncoder()
marital_status	 = LabelEncoder()


relationship = LabelEncoder()
race	 = LabelEncoder()
sex = LabelEncoder()
native_country = LabelEncoder()
income = LabelEncoder()
occupation = LabelEncoder()
income =  LabelEncoder()
```

```python
inputs['workclass_n'] = workclass.fit_transform(inputs['workclass'])
inputs['education_n'] = education.fit_transform(inputs['education'])
inputs['marital_status_n'] = marital_status.fit_transform(inputs['marital_status'])
inputs['relationship_n'] = relationship.fit_transform(inputs['relationship'])
inputs['race_n'] = race.fit_transform(inputs['race'])
inputs['sex_n'] = sex.fit_transform(inputs['sex'])
inputs['occupation_n'] = occupation.fit_transform(inputs['occupation'])
inputs['native_country_n'] = native_country.fit_transform(inputs['native_country'])


inputs_n = inputs.drop(['workclass','education','marital_status','relationship','race','sex','native_country','occupation'],axis='columns')
inputs_n



```

```python
from sklearn.model_selection import train_test_split
train_X,test_X,train_y,test_y= train_test_split(inputs_n, target, test_size=0.1,random_state=None)


```

```python
from sklearn import tree
model = tree.DecisionTreeClassifier()
```

```python
model.fit(train_X, train_y)
```

```python
from sklearn.metrics import accuracy_score
acurracy=model.score(test_X,test_y)
print(acurracy)
```
